# NumPy


NumPy is a library for working with arrays of numbers. It is very useful when you want to do the same operation on many values at once.


````python
import numpy as np
````

## Arrays

A NumPy array is a bit like a list, but it is designed for numerical work.


````python
a = np.array([1, 2, 3, 4])
a
````

````python
type(a)
````

You can still access elements by index.


````python
a[0], a[-1]
````

## Array operations

A very nice feature is that operations happen element-by-element.


````python
a + 10
````

````python
a * 2
````

````python
a ** 2
````

This also works with two arrays of the same shape.


````python
b = np.array([5, 6, 7, 8])
a + b
````

## Some useful methods

NumPy arrays come with many helpful methods and functions.


````python
a.sum()
````

````python
a.mean()
````

````python
np.linspace(0, 1, 5)
````
