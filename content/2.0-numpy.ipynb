# NumPy


NumPy is a library for working with arrays of numbers. It is very useful when you want to perform operations on numerical data.

At first sight, this might sound a lot like Python lists. And indeed, lists are often a perfectly good place to start. But NumPy arrays are much better suited for numerical work.

With plain Python lists, many operations need an explicit loop. If you want to add `10` to every element, or multiply two collections element-by-element, you usually have to write that logic yourself. NumPy lets you express these operations directly, and it is much faster as well.

So a good rule of thumb is: if you just want a general-purpose collection of things, a list is great. If you want to do mathematics on many numbers, NumPy is the better tool.

````python
import numpy as np
````

## Arrays

A NumPy array is a bit like a list, but it is designed for numerical work.


````python
a = np.array([1, 2, 3, 4])
a
````

````python
type(a)
````

You can still access elements by index.


````python
a[0], a[-1]
````

## Array operations

A very nice feature is that operations happen element-by-element.


````python
a + 10
````

````python
a * 2
````

````python
a ** 2
````

This also works with two arrays of the same shape.


````python
b = np.array([5, 6, 7, 8])
a + b
````

## Some useful methods

NumPy arrays come with many helpful methods and functions. There's to many to list, buy you can see then with `np.<press tab>` or `<array_name>.<press tab>`.

````python
a.sum()
````

````python
a.mean()
````

````python
np.linspace(0, 1, 5)
````

NumPy also has a `random` submodule, which provides powerful tools for random number generation. This is very useful when you want to generate random numbers, random samples, or random arrays for testing and examples.

However, random number generation is a topic where it is important to use the tools correctly. We will not go into detail here, but it is worth remembering that there is more to it than simply calling a random function.


````python
np.random.randint(0, 10, size=5)
````

## Multi-dimensional arrays

NumPy arrays can also have more than one dimension. A two-dimensional array is a bit like a table with rows and columns.


````python
m = np.array([
    [1, 2, 3],
    [4, 5, 6],
])
m
````

````python
m.shape
````

The shape tells us the size of each dimension. Here we have `2` rows and `3` columns.


````python
m[0, 0]
````

````python
m[1, 2]
````

If you give only one index, NumPy returns a whole row.


````python
m[0]
````

## Slicing

Slicing works much like it does for lists. It lets you select part of an array.


````python
a[1:3]
````

````python
a[:2]
````

````python
a[::2]
````

With a two-dimensional array, you can slice rows and columns separately.


````python
m[:, 1]
````

````python
m[0:2, 1:3]
````

The `:` means "take everything" along that dimension.


One important difference from Python lists is that NumPy slices usually do not copy the data. They give you a *view* into the same array.

This means that changing the slice can also change the original array. With ordinary Python lists, a slice usually creates a new list instead.


````python
numbers = [1, 2, 3, 4]
part = numbers[1:3]
part[0] = 999
numbers, part
````

````python
a = np.array([1, 2, 3, 4])
part = a[1:3]
part[0] = 999
a, part
````

## Fancy indexing

Sometimes you do not want a continuous slice. Fancy indexing lets you pick out several specific elements at once.


````python
a[[0, 2, 3]]
````

You can also use a boolean array to select values matching a condition.


````python
a > 2
````

````python
a[a > 2]
````

This is very useful for filtering data.

You can also select elements by selecting row and column indices separately. For example, to pick `(0, 2)` and `(1, 0)` you can use this.

````python
m[[0, 1], [2, 0]]
````
