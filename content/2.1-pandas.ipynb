# pandas


pandas is a library for working with tables of data. It is especially useful when your data has named columns.


````python
import pandas as pd
````

## DataFrames

The main pandas object is the `DataFrame`, which looks a lot like a spreadsheet.


````python
df = pd.DataFrame({
    "name": ["Ada", "Grace", "Linus", "Guido"],
    "score": [9, 8, 7, 10],
    "passed": [True, True, False, True],
})
df
````

Columns can be selected by name.


````python
df["score"]
````

## Adding a new column

You can create a new column from old ones.


````python
df["score_times_two"] = 2 * df["score"]
df
````

## Filtering rows

A comparison gives a boolean series, and that can be used to filter the table.


````python
df["score"] >= 9
````

````python
df[df["score"] >= 9]
````

## Useful summaries

pandas makes it easy to compute quick summaries.


````python
df["score"].mean()
````

````python
df.describe()
````
