# pandas


pandas is a library for working with tables of data. It is especially useful when your data has named columns.

At first sight, this might sound a bit like NumPy, since NumPy can also store data in arrays. But pandas is often a better choice when your data is naturally organised as a table, with rows representing entries and columns representing different kinds of information.

For example, if each row is a student and the columns are things like `name`, `score`, and `passed`, then pandas gives you a very natural way to work with that data. You can select columns by name, filter rows using conditions, and compute summaries very easily.

A good rule of thumb is: if your data is mostly numerical and you care mainly about array operations, NumPy is often the better tool. If your data is tabular and the column names matter, pandas is often more convenient.


````python
import pandas as pd
````

## DataFrames

The main pandas object is the `DataFrame`, which looks a lot like a spreadsheet.


````python
df = pd.DataFrame(
    {
        "name": ["Ada", "Grace", "Linus", "Guido"],
        "score": [9, 8, 7, 10],
        "passed": [True, True, False, True],
    }
)
df
````

Columns can be selected by name.


````python
df["score"]
````

## Adding a new column

You can create a new column from old ones.


````python
df["score_times_two"] = 2 * df["score"]
df
````

## Filtering rows

A comparison gives a boolean series, and that can be used to filter the table.


````python
df["score"] >= 9
````

````python
df[df["score"] >= 9]
````

## Useful summaries

pandas makes it easy to compute quick summaries.


````python
df["score"].mean()
````

````python
df.describe()
````
