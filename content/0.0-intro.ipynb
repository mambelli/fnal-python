## What is Python?

It is a high-level, general-purpose programming language that focuses on simplicity and user-friendliness.

It is interpreted, meaning that each line is read and executed at runtime. This is as opposed to compiled languages like C/C++, which get fully translated into machine instructions before they run.

## Why use Python?

Python has become the go-to language for scripting or for glueing computations or pieces of software together. The simplicity and extensive ecosystem make it very easy for you to tackle any task you might want.

In particular, for data analysis and machine learning, Python has become the standard language.

## Installing Python

There's plenty of options, depending on your operating system. The official installations instructions are [here](https://www.python.org/downloads/).

One well established option is using [Mamba](https://conda-forge.org/download/) a faster and more open alternative to [Anaconda](https://www.anaconda.com/download). It allows you to easily create separate environments for each of your projects and install almost any dependency you might need.

## Python Myths

- **Python is easy**

  The syntax in Python is indeed very easy and intuitive. However, that doesn't mean that it is easy to write good code. You still need to understand what's going on and use your tools correctly.

- **Python is slow**

  Pure Python is indeed slow. However, Python provides you with tools to wrap compiled code and many people have already written efficient packages that perform complex computations. If you use the right tools for the task, your Python code will be very fast.
