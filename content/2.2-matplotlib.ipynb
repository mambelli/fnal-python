# matplotlib


matplotlib is a library for making plots. It is one of the most common tools in Python for turning data into figures.

It gives you a lot of control over what your plots look like, which is very useful, although it can sometimes feel a little verbose at first. We will start with a few simple examples.


````python
import matplotlib.pyplot as plt
````

## A line plot

The simplest plot is often a line through a set of points.


````python
x = [0, 1, 2, 3]
y = [0, 1, 4, 9]
plt.plot(x, y)
plt.show()
````

We can add labels and a title to make the plot easier to read.


````python
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("A simple curve")
plt.show()
````

## Other plot types

matplotlib can also make scatter plots and histograms.


````python
plt.scatter([1, 2, 3, 4], [4, 1, 3, 2])
plt.show()
````

````python
values = [1, 1, 2, 2, 2, 3, 4, 4, 5]
plt.hist(values)
plt.show()
````

## Working with figure and axes

A common pattern is to create a figure and axes object, and then draw on the axes.


````python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("The same curve again")
plt.show()
````

## matplotlib wrappers

These tools are often very convenient, but it is still useful to know matplotlib, since many of them rely on it underneath.
